```python
def cartisan_2D_to_spherical(X):
    r = np.linalg.norm(X, axis=0)
    theta = np.apply_along_axis(lambda x: np.arctan2(x[1],x[0]), 0, X)
    return np.vstack((r, theta))
```

```python
def cartisan_3D_to_spherical(xyz):
    pts = np.zeros(xyz.shape)
    xy = xyz[0,:]**2 + xyz[1,:]**2
    pts[0,:] = np.sqrt(xy + xyz[2,:]**2)
    pts[1,:] = np.arctan2(np.sqrt(xy), xyz[2,:]) # for elevation angle defined from Z-axis down
    #pts[1,:] = np.arctan2(xyz[:,2], np.sqrt(xy)) # for elevation angle defined from XY-plane up
    pts[2,:] = np.arctan2(xyz[1,:], xyz[0,:])
    return pts
```

```python
def make_spherical_voxel_grid(G, SG, L, N, rot=None):
    """
    Discretise the evaluations of the field over the reciprocal space
    in the circle around origin with radius pi/L * N * sqrt(d) =  pi/delta_L * sqrt(d)
    where delta_L = L/N and N is the number of voxels along each axis
    
    Thus transforming G linearly such that its spherical coordinates are mapped to [0, N]^d.
    Then it the resulting vectors are rounded down to form the final grid.
    """
    
    d, n = G.shape
    absSG = np.abs(SG)
    voxel_width = np.pi / L * np.sqrt(d)
    radius = voxel_width * N # = 2*pi/delta_L * sqrt(d)
    
    if rot is None:
        rot = np.eye(d) # no rotation
    
    RG = rot.dot(G)
    
    grid_shape = (0,)
    if d == 2:
        grid_shape = (N,N)
        
        RG_spherical = cartisan_2D_to_spherical(RG)
        
        r = RG_spherical[0,:] # in [0, infty]
        phi = RG_spherical[1,:] # in [-pi, pi]

        I = np.floor(np.vstack((
            r / voxel_width,
            (phi + np.pi) / (2*np.pi/N)
        ))).astype(np.int)
    if d == 3:
        grid_shape = (N,N,N)
        
        RG_spherical = cartisan_3D_to_spherical(RG)
        
        r = RG_spherical[0,:] # in [0, infty]
        theta = RG_spherical[1,:] # in [0, pi]
        phi = RG_spherical[2,:] # in [-pi, pi]

        I = np.floor(np.vstack((
            r / voxel_width,
            theta / (np.pi/N),
            (phi + np.pi) / (2*np.pi/N)
        ))).astype(np.int)
    
    
    grid = np.zeros(grid_shape)

    in_grid = np.all((0 <= I) & (I < N), axis=0)
    I = I[:,in_grid]
    absSG = absSG[in_grid]
    
    for i in range(I.shape[1]):
        index = I[:,i]
        if d == 2:
            grid[index[0], index[1]] += absSG[i] # rows = y-axis, cols = x-axis
        if d == 3:
            grid[index[0], index[1], index[2]] += absSG[i]

    return grid
```

```python
R = get_2D_rotation_matrix(np.pi/4) # 45 degress

# final step:
# place the computed values on a N x N grid
descriptor = adapt_to_voxel_grid(G, SG, L, N, rot=R)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
plt.imshow(descriptor, origin='lower')
plt.xticks(np.arange(0,N)+0.5)
plt.yticks(np.arange(0,N)+0.5)
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.grid(True)
```

## Real molecule example

```python
import pandas as pd
o_sr_ti = pd.read_pickle("OSrTi.pkl")

from mol_tools import *
```

```python
mol = o_sr_ti.loc[1]
mol.compound
```

```python
# all atoms in one plane
%matplotlib widget
plot_3D_crystal(mol)
```

```python
A = calc_basis(mol.geometry)
A
```

```python
# perform rotation to eliminate dimesion for better visualisation
v = (A[:,0] + A[:,1])[[0,1]]
v = v / np.linalg.norm(v)
a = np.arccos(v[0])
rot2 = get_2D_rotation_matrix(a)
rot = np.eye(3)
rot[0:2,0:2] = rot2
rot
```

```python
rot_coords = rot.T.dot(calc_cartesian_positions(A, mol.positions_fractional).T).T
rot_coords
```

```python
# new coordinates
mus2 = rot_coords[:, [0,2]]
mus2
```

```python
# new unit cell
az = rot.T.dot(A[:,2])[[0,2]]
A2 = np.array([
    [np.linalg.norm(A[:,0] + A[:,1]), az[0]], 
    [0., az[1]]
])
A2
```

```python
sigma = 0.1
gauss = get_nD_gaussian(A2, mus2, sigma, 3)
R, SR = bravais_lattice(gauss, A2, n=201)
plot_2D_realspace_lattice(A2, R, SR)
```

```python
L = 12.8
N = 32

width = 2 * np.pi / L * N

mx, my = get_mesh_coords(A2, width)

B, G, SG = reciprocal_lattice_gaussian(A2, mus2, sigma, mx, my, mz)

plot_2D_reciprocal_lattice(B, mx, my, G, SG, L=L, N=N)
```

```python
R = get_2D_rotation_matrix(np.pi/4) # 45 degress

# final step:
# place the computed values on a N x N grid
descriptor = adapt_to_voxel_grid(G, SG, L, N, rot=np.eye(2))

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
plt.imshow(descriptor, origin='lower')
plt.xticks(np.arange(0,N)+0.5)
plt.yticks(np.arange(0,N)+0.5)
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.grid(True)
```
