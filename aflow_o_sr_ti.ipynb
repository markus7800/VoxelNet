```python
fig = plt.figure()
ax = Axes3D(fig)

no = r.composition[0]
ns = r.composition[1]
nt = r.composition[2]

colors = ["red"]*no + ["green"]*ns + ["blue"]*nt

ax.scatter(poss[:,0], poss[:,1], poss[:,2], alpha=1, c=colors, s=25)
for i in range(3):
    ax.plot([0, A[0,i]], [0,A[1,i]], [0,A[2,i]])
plt.show()
```
