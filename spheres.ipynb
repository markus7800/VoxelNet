# Test on Spheres

```python
def spheres(r, v):
    x = v[0]
    y = v[1]
    
    cx = np.round(x)
    cy = np.round(y)
    
    if (x-cx)**2 + (y-cy)**2 < r**2:
        return 1.
    else:
        return 0.
    
def get_spheres_field(r, offset):
    def F(v):
        v += offset
        return spheres(r, v)
    return F
```

```python
centered_sphere = get_spheres_field(0.25, offset=np.array([0.5,0.5]))
corner_spheres = get_spheres_field(0.25, offset=np.array([0.,0.]))
```

## One Centered Sphere in Unit Cell

```python
A = np.eye(2)
xi, sr = bravais_lattice(centered_sphere, A)
plot_2D_bravais_lattice(A, xi, sr)
```

```python
B, mi, sg = reciprocal_lattice(centered_sphere, A, g0=-16, g1=16)
plot_2D_reciprocal_lattice(B, mi, sg)
```

## One Sphere at Corner of Unit Cell

```python
xi, sr = bravais_lattice(corner_spheres, A)
plot_2D_bravais_lattice(A, xi, sr)
```

```python
B, mi, sg = reciprocal_lattice(corner_spheres, A, g0=-16, g1=16)
plot_2D_reciprocal_lattice(B, mi, sg)
```
